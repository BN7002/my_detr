```python
import pandas as pd
import json
import matplotlib.pyplot as plt

with open('coco/annotations/instances_train2017.json', 'r') as f:
    data = json.load(f)

annotations = pd.DataFrame(data['annotations'])
category_counts = annotations['category_id'].value_counts().sort_index()

print(f"Number of unique classes: {annotations['category_id'].nunique()}")

plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
plt.xlabel('Category ID')
plt.ylabel('Number of Occurrences')
plt.title('Occurrences of Each Category ID in COCO train2017')
plt.show()
```
